```python
#selecting 'STATE/UT' and 'Total IPC CRIMES' columns from 'a' dataframe
totalcrime=a[['STATE/UT','DISTRICT','TOTAL IPC CRIMES']]
totalcrime
```

As the **'TOTAL IPC CRIMES'** column is **district-wise**, we can **add all crimes** in **corresponding district** and **assign** them **according** to the **'STATE/UT'.**

```python
#filtering the DataFrame based on the 'TOTAL' value in the 'DISTRICT' column, as it represents the total value for the remaining districts.
totalcrime=totalcrime[totalcrime['DISTRICT']=='TOTAL']

totalcrime=totalcrime.groupby('STATE/UT')['TOTAL IPC CRIMES'].sum()#adding values to the corresponding 'STATE/UT'
totalcrime
```

Theese are the total crimes of each state/ut.

```python
#merging data based on 'STATE/UT'
df=pd.merge(df,totalcrime,on='STATE/UT')
df
```

```python
#selecting all columns except for 'YEAR','OTHER IPC CRIMES' and 'TOTAL IPC CRIMES' columns  from 'a' data frame
selected_columns=a.drop(['YEAR','OTHER IPC CRIMES', 'TOTAL IPC CRIMES'],axis=1)

#filtering the DataFrame based on the 'TOTAL' value in the 'DISTRICT' column, as it represents the total value for the remaining districts.
selected_columns=selected_columns[selected_columns['DISTRICT']=='TOTAL']
```

```python
crime=pd.DataFrame()#creating empty data frame

columns=selected_columns.columns.to_list()#storing columns of 'selected_columns' as a list

#removing unwanted columns
columns.remove('STATE/UT')
columns.remove('DISTRICT')

for i in columns:#iterating through 'columns'
    
    crime[i]=selected_columns.groupby('STATE/UT')[i].sum()#creating new column in 'crime' and adding all values present in 'i' column


crime.reset_index(inplace=True)#reseting index 
crime
```

From this data frame, we can assess the **type of crime in each state** by **assigning** the **category** with the **highest value**.

```python
selected_data=crime.drop('STATE/UT',axis=1)#dropping 'STATE/UT' column and storing remaining data 

type_c=[]#creating empty list

for i in range(0,34):#running loop for 34 times
    type_c.append(selected_data.iloc[i].idxmax())  #appending column to 'type_c' according to max value present in each row
    
#creating new data frame
type_of_crime=pd.DataFrame({'STATE/UT':crime['STATE/UT'],'TYPE OF CRIME':type_c})
type_of_crime
```

We can **merge this data frame** with **'df' data frame**.

```python
df=pd.merge(df,type_of_crime,on='STATE/UT')
df
```

Data is merged successfully.
